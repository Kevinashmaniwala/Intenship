```python
# --- Income માટે Outlier Treatment (IQR Method) ---
Q1_inc = df['Income'].quantile(0.25)
Q3_inc = df['Income'].quantile(0.75)
IQR_inc = Q3_inc - Q1_inc

lower_inc = Q1_inc - 1.5 * IQR_inc
upper_inc = Q3_inc + 1.5 * IQR_inc

df['Income'] = df['Income'].clip(lower=lower_inc, upper=upper_inc)

# --- Debt માટે Outlier Treatment ---
Q1_debt = df['Debt'].quantile(0.25)
Q3_debt = df['Debt'].quantile(0.75)
IQR_debt = Q3_debt - Q1_debt

lower_debt = Q1_debt - 1.5 * IQR_debt
upper_debt = Q3_debt + 1.5 * IQR_debt


df['Debt'] = df['Debt'].clip(lower=lower_debt, upper=upper_debt)

```
