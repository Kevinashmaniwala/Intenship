 


## 📄 Dataset Report: Women’s Clothing E-Commerce Reviews

### 1. Objective of the Report

The purpose of this report is to analyze customer reviews of women’s clothing products from an e-commerce platform to understand:

* Customer satisfaction
* Product recommendations
* Ratings and feedback patterns
* Demographic influence (Age)

---

### 2. Dataset Overview

* **Total Records:** 23,486
* **Total Columns:** 11
* **Type of Data:** Structured (CSV file)

This dataset contains customer review details such as rating, review text, age, product category, and recommendation status.

---

### 3. Column Description

| Column Name             | Description                               |
| ----------------------- | ----------------------------------------- |
| Unnamed: 0              | Index column                              |
| Clothing ID             | Unique ID for each product                |
| Age                     | Age of the customer                       |
| Title                   | Review title (may contain missing values) |
| Review Text             | Customer’s written review                 |
| Rating                  | Rating given by customer (1 to 5)         |
| Recommended IND         | 1 = Recommended, 0 = Not Recommended      |
| Positive Feedback Count | Number of positive votes on review        |
| Division Name           | High-level product division               |
| Department Name         | Product department                        |
| Class Name              | Specific product category                 |

---

### 4. Data Quality Observations

* Some **missing values** in:

  * Review Title
  * Review Text
  * Division, Department, and Class Name
* Rating, Age, and Recommendation columns have **no missing values**
* Index column (`Unnamed: 0`) is not useful for analysis and can be removed

---

### 5. Key Insights

#### ⭐ Ratings Analysis

* Most ratings are **4 and 5**, indicating high customer satisfaction
* Lower ratings (1–2) exist but are comparatively fewer

#### 👍 Recommendation Analysis

* Majority of customers **recommend the products**
* High ratings strongly correlate with recommendation = 1

#### 👥 Age Analysis

* Customers from **middle age groups** are the most active reviewers
* All age groups participate in reviewing

#### 🛍 Product Category Insights

* Most reviews belong to:

  * **Dresses**
  * **Tops**
  * **Bottoms**
* Certain departments receive consistently higher ratings

#### 💬 Feedback Count

* Reviews with **higher positive feedback counts** are usually detailed and positive
* Helpful reviews influence other customers

---

### 6. Business Use Cases

* Identify **top-rated product categories**
* Improve products with **low ratings**
* Use review text for **sentiment analysis**
* Enhance marketing by targeting **highly recommended items**

---

### 7. Conclusion

The dataset shows that customers are generally satisfied with women’s clothing products. Ratings and recommendations indicate strong product performance, while review texts provide valuable insights into customer preferences and concerns.

    Regression:

```python
tfidf = TfidfVectorizer(max_features=1500, stop_words='english')
X_text = tfidf.fit_transform(data['Review Text']).toarray()

X_user = data[['Age', 'Class Name', 'Positive Feedback Count', 'Department Name']]
X_final = np.hstack([X_user, X_text]) # merge both sell for used hstack

y = data['Rating']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_final,y,test_size=0.4,random_state=42)
```
